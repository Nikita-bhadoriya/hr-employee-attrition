```python
#Employee Attrition Count
sns.countplot(x='Attrition',data=data)
plt.title('Employee Attrition Count')
plt.show()
```

```python
# Age Distribuation
sns.histplot(data['Age'],bins=30,kde=True)
plt.title('Age Distribuation')
plt.show()
```

```python
# Department count
sns.countplot(x='Department',data=data)
plt.title('Department Count')
#plt.xticks(rotation=2)
plt.show()
```

```python
#Attrition vs overtime
sns.countplot(x='OverTime',hue='Attrition',data=data)
plt.title('Attrition by overtime')
plt.show()
```

```python
# Attrition vs Job Satisfaction
sns.boxplot(x='Attrition',y='JobSatisfaction',data=data)
plt.title('Attrition vs Jobsatisfation')
plt.show()
```

```python
# Monthly Income Distribution by Attrition
sns.boxplot(x='Attrition',y='MonthlyIncome',data=data)
plt.title('Attrition vs monthly income')
plt.show()
```

```python
# corrletion heatmap
plt.figure(figsize=(15,10))
sns.heatmap(data.corr(numeric_only=True),cmap='Reds')
plt.title('Correlation Heatmap')
plt.show()
```
